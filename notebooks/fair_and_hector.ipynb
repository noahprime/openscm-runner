# Using OpenSCM
This notebook is intended to give an example of how to use OpenSCM-Runner to easily run multiple simple climate models (in this case FaIR and Hector) on an ensemble of the same future emission pathways, and extract results.

```python
# Import Packages
import os.path

import matplotlib.pyplot as plt
import scmdata
from openscm_runner.run import run
```

```python
# Emission Pathways
input_emissions = scmdata.ScmRun(
    os.path.join(
        "..", "tests", "test-data", "rcmip_scen_ssp_world_emissions.csv"
    ),
    lowercase_cols=True,
)

input_emissions.head(30)
```

## Running the SCMs
Here we run FaIR and Hector for each scenario in the input emissions data. In this case we are running both models using their respective default configuration, though users can just as easily do multiple runs with modified model parameters.

```python
x = run(
    climate_models_cfgs={
        # Climate models, each with a list of configuarations
        "FAIR": [{}],
        "HECTOR": [{}]
    },
    scenarios=input_emissions,
    # Variables we want to have data returned for
    output_variables=(
        "Surface Air Temperature Change",
        "Atmospheric Concentrations|CO2",
        "Effective Radiative Forcing",
        "Effective Radiative Forcing|CO2",
        "Effective Radiative Forcing|Aerosols",
        "Effective Radiative Forcing|Aerosols|Direct Effect|BC",
        "Effective Radiative Forcing|Aerosols|Direct Effect|OC",
        "Effective Radiative Forcing|Aerosols|Direct Effect|SOx",
        "Effective Radiative Forcing|Aerosols|Direct Effect",
        "Effective Radiative Forcing|Aerosols|Indirect Effect",
    ),
)
```

### Plots
We get to use `scmdata`'s built in filtering and plotting functionality to visualize results from the ensemble runs.

```python
# Plot Aesthetics
plot_kwargs = dict(
    #quantiles_plumes=[((0.05, 0.95), 0.5), ((0.5,), 1.0)],
    quantile_over="run_id",
    hue_var="scenario",
    style_var="model",
    style_label="IAM",
    time_axis="year",
)
```

#### Surface Air Temperatre Change

```python
%%capture --no-display
ax = plt.figure(figsize=(12, 7)).add_subplot(111)
x.filter(variable="Surface Air Temperature Change", climate_model = 'Hector').plumeplot(
    ax=ax, **plot_kwargs
)
ax.axvline(2015)
ax.set_title(f'Hector, Surface Air Temperature Change')
```

```python
%%capture --no-display
ax = plt.figure(figsize=(12, 7)).add_subplot(111)
x.filter(variable="Surface Air Temperature Change", climate_model = 'FaIRv1.6.4').plumeplot(
    ax=ax, **plot_kwargs
)
ax.axvline(2015)
ax.set_title(f'FaIR, Surface Air Temperature Change')
```

#### Atmospheric CO2 Concentrations

```python
%%capture --no-display
ax = plt.figure(figsize=(12, 7)).add_subplot(111)
x.filter(variable="Atmospheric Concentrations|CO2", climate_model = 'Hector').plumeplot(
    ax=ax, **plot_kwargs
)
ax.axvline(2015)
ax.set_title(f'Hector, Atmospheric Concentrations|CO2')
```

```python
%%capture --no-display
ax = plt.figure(figsize=(12, 7)).add_subplot(111)
x.filter(variable="Atmospheric Concentrations|CO2", climate_model = 'FaIRv1.6.4').plumeplot(
    ax=ax, **plot_kwargs
)
ax.axvline(2015)
ax.set_title(f'FaIR, Atmospheric Concentrations|CO2')
```

### Example Summary Statistics
Again, `scmdata` functions like `filter` allow you to extract results for any returned variable of interest, for the given years, scenario, SCM or IAM.

```python
fair_ssp126 = x.filter(
    variable = "Surface Air Temperature Change", 
    climate_model = 'FaIRv1.6.4',
    year = 2100,
    scenario=["ssp126"]
    ).values.max()

fair_ssp585 = x.filter(
    variable = "Surface Air Temperature Change", 
    climate_model = 'FaIRv1.6.4',
    year = 2100,
    scenario=["ssp585"]
    ).values.max()

hector_ssp126 = x.filter(
    variable = "Surface Air Temperature Change", 
    climate_model = 'Hector',
    year = 2100,
    scenario=["ssp126"]
    ).values.max()

hector_ssp585 = x.filter(
    variable = "Surface Air Temperature Change", 
    climate_model = 'Hector',
    year = 2100,
    scenario=["ssp585"]
    ).values.max()

f'In 2100, the maximum surface air temperature change from the set of FaIR runs of SSP 126 and SSP 585, ' +\
f'were {fair_ssp126:.2f} and {fair_ssp585:.2f} C respectively. For Hector, {hector_ssp126:.2f} and ' +\
f'{hector_ssp585:.2f} C respectively.'
```
